```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read the CSV file containing gas prices
gas = pd.read_csv('gas_prices.csv')

# Set the figure size for the plot
plt.figure(figsize=(10, 5))

# Add a title to the plot
plt.title("Gas prices through the years (in USD)", fontdict={'fontsize': 20, 'fontweight': 'bold'})

# Plot gas prices for different countries
plt.plot(gas.Year, gas.USA, 'b.-', label='USA')
plt.plot(gas.Year, gas.Japan, 'r*-', label='Japan')
plt.plot(gas.Year, gas.Australia, 'y^-', label='Australia')

# Label the axes
plt.xlabel("Years")
plt.ylabel("Price in USD")

# Set x-ticks and y-ticks
plt.xticks(gas.Year[::2])
plt.yticks([1, 2, 3, 4, 5, 6])

# Display the legend
plt.legend()

# Save the plot as a PNG file
plt.savefig('Gas_prices_usd.png', dpi=300)

# Show the plot
plt.show()
```
